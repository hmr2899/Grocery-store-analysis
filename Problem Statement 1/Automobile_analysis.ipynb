```python
rfm= RFM_data.groupby('CUSTOMERNAME').agg({'ORDERDATE': lambda date: (PRESENT - date.max()).days,
                                        'ORDERNUMBER': lambda num: len(num),
                                        'SALES': lambda price: price.sum()})
```

```python
rfm.columns=['Recency','Frequency','Monetary']
```

```python
rfm['Recency'] = rfm['Recency'].astype(int)
```

```python
rfm.head()
```

```python
quantiles = rfm.quantile(q=[0.25,0.5,0.75])
quantiles
```

```python
quantiles = quantiles.to_dict()
```

```python
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
```

```python
rfm_segmentation = rfm
```

```python
rfm_segmentation['R_Quartile'] = rfm_segmentation['Recency'].apply(RScore, args=('Recency',quantiles,))
```

```python
rfm_segmentation['F_Quartile'] = rfm_segmentation['Frequency'].apply(FMScore, args=('Frequency',quantiles,))
```

```python
rfm_segmentation['M_Quartile'] = rfm_segmentation['Monetary'].apply(FMScore, args=('Monetary',quantiles,))
```

```python
rfm_segmentation.head()
```

```python
rfm_segmentation['RFMScore'] = rfm_segmentation.R_Quartile.map(str) \
                            + rfm_segmentation.F_Quartile.map(str) \
                            + rfm_segmentation.M_Quartile.map(str)
```

```python
rfm_segmentation.head()
```

```python
# Q1. Who are the best Customers?

rfm_segmentation[rfm_segmentation['RFMScore']=='444'].sort_values('Monetary', ascending=False).head(6)
```
